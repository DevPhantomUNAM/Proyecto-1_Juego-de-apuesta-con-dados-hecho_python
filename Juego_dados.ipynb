```python
def definirApuesta(args):
    
    ganancia = [0]    
    
    resultado = tirar_dados()
    resultadoSuma = sum(resultado)
    par = sum(resultado)%2
    
    print(resultadoSuma)
    
    #Pares e Impares
    if args[0][1] > 0:
        if par == 1 and args[0][1] == 1:
            ganancia.append(args[0][0]*2)
        elif par == 0 and args[0][1] == 2:
            ganancia.append(args[0][0]*2)
            
    
    
    #Colores
    if args[1][1] > 0:
        if par == 1 and args[1][1] == 1:
            ganancia.append(args[1][0]*2)
        elif par == 0 and args[1][1] == 2:
            ganancia.append(args[1][0]*2)
            
    if sum(resultado)%2 == 0:
        print("Salio par ")
        print("Rojo")
    else: 
        print("Salio inpar")
        print("Negro")
            
    #Columnas  
    
    columna_1 = [4,7,10,13,16]
    columna_2 = [5,8,11,14,17]
    columna_3 = [6,9,12,15,18]
    
    if args[2][1] > 0:
        if resultadoSuma in columna_1:
            ganancia.append(args[2][1] * 33)
        elif resultadoSuma in columna_1:
            ganancia.append(args[2][2] * 33)
        elif resultadoSuma in columna_1:
            ganancia.append(args[2][3] * 33)
        
        if resultadoSuma in columna_1:
            print("Columna 1")
        if resultadoSuma in columna_2:
            print("Columna 2")
        if resultadoSuma in columna_3:
            print("Columna 3")

    return sum(ganancia)
```
